```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
m = 100
x = np.random.rand(m, 1)
y = 4 + 3 * x + np.random.randn(m, 1) * 0.1
```

```python
# x
```

```python
plt.figure(figsize=(5, 3))
plt.scatter(x, y, color='blue', s=5)
plt.show()
```

```python
w = 0
b = 0
lr = 0.1
n = 100
# iteration = [10, 20, 30, 40, 50, 60, 70 , 80, 90]
iteration = 200
lst = []
```

```python
for i in range(iteration):
    y_pred = w * x + b
    error = y_pred - y
    loss = (1/n) * np.sum(error ** 2)
    
    lst.append(loss)
    
    dw = (2/n) * np.sum(error * x)
    db = (2/n) * np.sum(error)
    
    w = w - lr * dw
    b = b - lr * db
```

```python
print(w)
```

```python
print(b)
```

```python
plt.figure()
plt.scatter(x, y, color='blue')
plt.plot(x, w * x + b, color='red')
plt.show()
```

```python
plt.plot(range(iteration), lst)
plt.show()
```
